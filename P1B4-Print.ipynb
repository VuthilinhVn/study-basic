# Cú pháp:

print(object(s), sep=separator, end=end, file=file, flush=flush)

where:

    - object(s): đối tượng, dữ liệu đc viết
    - sep: ngăn cách
    - end: kết thúc
    - file: tên tập tin
    - flush: đẩy dữ liệu
