# Thư viện Math
- math.ceil(x) : giá trị trần của số nguyên, số nguyên nhỏ nhất lớn hơn hoặc bằng x => tròn lên
- math.fabs(x): giá trị tuyệt dodói của x, trả về dữ liệu kiểu float
- abs(x): giá trị tuyệt đối của x, trả về int
- math.floor(x): giá trị sàn của số nguyên , số nguyên nhỏ nhất hoặc bằng x => tròn xuống
- math.exp(x) : e lũy thừa x, e = 2.718281...
- math.log(x, base) -> math.log(x)= ln(x), math.log(x, 2)=log₂(x)=math.log2(x), math.log(x, 10)=log₁₀(x)=math.log10(x)
- math.pow(x, y): x lũy thừa y (x mũ y)
- math.sqrt(x): căn bậc 2 của x
- math.factorial(5) giai thừa
- math.gcd(12, 18) ước chung lớn nhất
- math.isclose(0.1 + 0.2, 0.3) so sánh số thực

# Giải phương trình bậc hai
y = ax^2+bx+c
a, b, c nhập từ bàn phím

```python
import math 

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a != 0:
    denta = (b**2 - 4*a*c)
    if denta <0:
        print("Ptrinh vô nghiệm!")
    elif denta == 0:
        print("Ptrinh có nghiẹm la: x1 = x2 = ", (-b)/(2*a))
    else:
        print("Ptrinh có 2 nghiem: ")
        print("x1 = ", (-b+math.sqrt(denta))/(2*a))
        print("x1 = ", (-b-math.sqrt(denta))/(2*a))

else:
    print("Không phải phường trình bậc 2!")
    
```

# Tính toán diện tích và chu vi của tam giác
- 3 điểm trên hệ trục tọa độ Oxy -> 3 điểm sẽ tạo thành tam giác nếu tổng của 2 cạnh lớn hơn cạnh còn lại
- Nếu tạo thành tam giác thì tính chu vi và diện tích của tam giác

```python
import math

xA = float(input("Nhập tọa độ x_a = "))
yA = float(input("Nhập tọa độ y_a = "))
xB = float(input("Nhập tọa độ x_b = "))
yB = float(input("Nhập tọa độ y_b = "))
xC = float(input("Nhập tọa độ x_c = "))
yC = float(input("Nhập tọa độ y_c = "))

AB = math.sqrt((xB-xA)**2 + (yB-yB)**2)
BC = math.sqrt((xC-xB)**2 + (yC-yB)**2)
AC = math.sqrt((xC-xA)**2 + (yC-yA)**2)

if (AB+AC > BC) and (AB+BC > AC) and (AC+BC > AB):
    print("Ba điểm tạo thành một tam giác!")
    chuvi = AB+BC+AC
    dienTich = math.sqrt(chuvi*(chuvi-AB)*(chuvi-AC)*(chuvi-BC))
    print(f"chu vi tam giác = {chuvi}, Diện tích tam giác = {dienTich}")
else:
    print("Ba điẻm không tạo thành tam giac!")

            
            
```

```python
# Kởi tạo list
emptyList = []
emptyList2 = list()

print(emptyList)
print(emptyList2)
```

```python
 # Tạo list mầu sắc
colors = ["red", "green", "black"]
print(colors)
```

```python
student = ["An", "Bình", "Ngân", "Vi", "Thắng"]

# lấy ra 1 bạn cuối cùng
last_student = student[-1]

# lấy ra bạn đầu tiên
first_student = student[0]

# lấy ra bạn ở vị trí thứ 2
second_student = student[1]

print("Bạn cuối cùng: ", last_student)
print("Bạn đầu tiên: ", first_student)
print("Bạn thứ hai:" , second_student)
```

```python
# lấy các bạn từ vị trí đầu tiên đến vị trí thứ 3
print(student[:3])

# lấy 3 bạn từ cuối lên
print(student[-3:])

# đảo list， ngoài ra còn có reverse

print(student[::-1])

new = student.reverse()
print(new)
```

```python
# lấy cách 
print(student[::2])
# copy list
new_list = student[:]
print(new_list)
```

```python
# Thêm phẩn tử vào cuối list
student.append("Thiên")
print(student)
print(new_list)
```

```python
# thêm phần tử vào vị trí cụ thể => list.insert(index, value)
student.insert(2, "Ngọc")
print(student)
```

```python
# tính độ dài list
length_student = len(student)
print(length_student)
```

```python
# hàm count
Ngoc_cnt = student.count("Ngọc")
print(Ngoc_cnt)
```

```python
student.remove("Ngọc")
```

```python
# xóa phần tử ra khỏi list bằng index => pop(index)
student.pop(0)
print(student)
```

# Vòng lặp For
- Cú pháp:
- 
  for biến in iterable:
  
    code
- where:
- (+) list
- (+) string
- (+) range(start, end, step)
- (+) tuple
- Lấy index và value:

  for i, s in enumerate(student):
  
    print(i, s)

```python
# tính tổng 0-> n
n = int(input("Nhập vào n = "))
tong = 0
for i in range(n+1):
        tong+=i
print(tong)
```

```python
student = ["Hi", "Ha", "He", "Hé"]
for ten in student:
    print(ten)
```

```python
for index, ten in enumerate(student):
    if index == 1:
        print(ten)
```

```python
# in ra các số chẵn từ 0-n
n = int(input("Nhập vào số n: "))
for i in range(n+1):
    if i%2==0 and i!=0:
        print(i)
```

# Kiểu dữ liệu String(chuỗi)
- Các chuỗi có thể cộng với nhau (+)
- chuỗi nhiều dòng sử dụng ''' (nhiều dòng here) '''
- kiểm tra 1 chuỗi có nằm bên trong 1 chuỗi khôngg? --> chuỗi 1 in chuỗi 2 (kết quả là true/false)
- Vieets hoa toàn bộ chữ đầu: string.capitalize()
- Viết hoa toàn bộ string: string.upper()
- Viết thường toàn bộ string: string.lower()
- Đếm có bao nhiêu giá trị chuỗi con : string.count("value") -> trả về số
- tìm chuỗi con string.find("value") -> trả về index
- thay thế  string.replace("Value 1", "Value sẽ thay thế")
- tách: string.split("ký tự muốn dùng để tách")
- 

```python
str = "hôm nay Trời đẹp quá!"
print(str)
# Viết hoa
print(str.upper())
# viêths thường
print(str.lower())
# viết hoa chữ đầu:
print(str.capitalize())
```

```python
print(str.count("nay"))
```

```python
print(str.find("nay"))
```

```python
str = str.replace("quá", "thế")
print(str)
```

```python
for i in str.split(" "):
    print(i)
```
