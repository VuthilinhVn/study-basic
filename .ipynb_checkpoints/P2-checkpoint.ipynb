# Thư viện Math
- math.ceil(x) : giá trị trần của số nguyên, số nguyên nhỏ nhất lớn hơn hoặc bằng x => tròn lên
- math.fabs(x): giá trị tuyệt dodói của x, trả về dữ liệu kiểu float
- abs(x): giá trị tuyệt đối của x, trả về int
- math.floor(x): giá trị sàn của số nguyên , số nguyên nhỏ nhất hoặc bằng x => tròn xuống
- math.exp(x) : e lũy thừa x, e = 2.718281...
- math.log(x, base) -> math.log(x)= ln(x), math.log(x, 2)=log₂(x)=math.log2(x), math.log(x, 10)=log₁₀(x)=math.log10(x)
- math.pow(x, y): x lũy thừa y (x mũ y)
- math.sqrt(x): căn bậc 2 của x
- math.factorial(5) giai thừa
- math.gcd(12, 18) ước chung lớn nhất
- math.isclose(0.1 + 0.2, 0.3) so sánh số thực

# Giải phương trình bậc hai
y = ax^2+bx+c
a, b, c nhập từ bàn phím

```python
import math 

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a != 0:
    denta = (b**2 - 4*a*c)
    if denta <0:
        print("Ptrinh vô nghiệm!")
    elif denta == 0:
        print("Ptrinh có nghiẹm la: x1 = x2 = ", (-b)/(2*a))
    else:
        print("Ptrinh có 2 nghiem: ")
        print("x1 = ", (-b+math.sqrt(denta))/(2*a))
        print("x1 = ", (-b-math.sqrt(denta))/(2*a))

else:
    print("Không phải phường trình bậc 2!")
    
```

# Tính toán diện tích và chu vi của tam giác
- 3 điểm trên hệ trục tọa độ Oxy -> 3 điểm sẽ tạo thành tam giác nếu tổng của 2 cạnh lớn hơn cạnh còn lại
- Nếu tạo thành tam giác thì tính chu vi và diện tích của tam giác

```python
import math

xA = float(input("Nhập tọa độ x_a = "))
yA = float(input("Nhập tọa độ y_a = "))
xB = float(input("Nhập tọa độ x_b = "))
yB = float(input("Nhập tọa độ y_b = "))
xC = float(input("Nhập tọa độ x_c = "))
yC = float(input("Nhập tọa độ y_c = "))

AB = math.sqrt((xB-xA)**2 + (yB-yB)**2)
BC = math.sqrt((xC-xB)**2 + (yC-yB)**2)
AC = math.sqrt((xC-xA)**2 + (yC-yA)**2)

if (AB+AC > BC) and (AB+BC > AC) and (AC+BC > AB):
    print("Ba điểm tạo thành một tam giác!")
    chuvi = AB+BC+AC
    dienTich = math.sqrt(chuvi*(chuvi-AB)*(chuvi-AC)*(chuvi-BC))
    print(f"chu vi tam giác = {chuvi}, Diện tích tam giác = {dienTich}")
else:
    print("Ba điẻm không tạo thành tam giac!")

            
            
```
