# Sự khác biệt giữa list và tuple
- List có thể cập nhật, xóa, thay đổi,... Mutable -> []
- Tuple thì k thể thay đổi đc (Immutable) -> () nằm trong ngoặc tròn
- 

```python
GioiTinh = ('Nam','Nữ')

# Các thao tác với Tuple
## truy cập
print(GioiTinh[0])

for x in GioiTinh:
    print(x)

# Cộng (gộp tuple)
#  nhân tuple -> nó sẽ lặp lại các phần tử trong tuple
print('Nam' in GioiTinh)

# độ dài tuple
x = len(GioiTinh)
# đếm số lần xuất hiện
print(GioiTinh.count('Nữ'))

# min, max dùng đc cho cả dữ liệu string hoặc numberic
# sum chỉ dùng đc với dữ liệu sum nhé
print(max(GioiTinh))
print(min(GioiTinh))

num = (1,2,3)
print(sum(num))

num2 = sorted(num, reverse = True)
print(num2)
```

# set
- Là một tập hợp không có thứ tự, k thể thay đổi immutable và không đc lập chỉ mục. Lưu ý: Các mục set là k thể thay đôi nhưng bạn có thể xóa các mục hoặc thêm các mục mới
-  giá trị k đc trùng nhau -> {}

```python
monHoc = {'Toán', 'Lý'}
print(monHoc)
```

```python
# Duyệt các phần tử của set
for i in monHoc:
    print(i)
```

```python
# thêm một phần tử vào set:
# (1) Dùng add
monHoc.add('Lịch Sử')
print(monHoc)

# (2) dùng update
monHoc.update({"Anh Văn", "Đàn Piana"})
print(monHoc)

# khi thêm list vào set thì chỉ để lại 1 giá trị duy nhất nhé -> k lặp lại
```

```python
# xóa phần tử:
# monHoc.remove("Toán")
# print(monHoc)

# sử dụng discard (có thì xóa, k có thì k lỗi, còn remove nếu k tìm thấy thì sẽ báo lỗi)
monHoc.discard("Lý")
print(monHoc)
```

```python
# xóa phần tử đầu tiên -> pop
monHoc.pop()
print(monHoc)
```

```python
# xóa sạch clear()
monHoc.clear()
print(monHoc)
```

```python
# xóa hẳn set đi -> del
del(monHoc)
print(monHoc)
```

# Bài tập rút thăm trúng thưởng
- Xây dựng chương trình rút thăm trúng thưởng với các chức năng sau:
- (+) thêm một mã số dự thưởng vào thùng
- (+) Xóa một mã số dự thưởng ra khỏi thùng
- (+) Quay số ngẫy nhiên lấy ra một mã số trúng thưởng => dùng random nhé
- 

```python
import random 
# khai báo thùng phiếu
thung = set()

while(True):
    print('-----------')
    print("Vui lòng lựa chọn chức năng!")
    print("1 - thêm một mã dự thưởng vào thùng")
    print("2 - xóa một mã dự thưởng ra khỏi thùng")
    print("3 - Quay số ngẫu nhiên lấy ra một mã số trúng thưởng")
    print("4 - Kết thúc chương trình")
    luaChon = int(input())
    if (luaChon == 1):
        sdt = input("Nhap vao sđt: ")
        thung.add(sdt)
    elif(luaChon == 2):
        delsdt = input("Nhap sđt cần xóa: ")4
        thung.discard(delsdt)
    elif(luaChon == 3):
        if len(thung) == 0:
            print("Thùng rỗng!")
        else:
            winner = random.choice(list(thung))
            print(f"Chúc mừng {winner} đã trúng thưởng")
            thung.discard(winner)
    elif(luaChon == 4):
        print("Thoat")
        break
    else:
        break
    
        
    
```

# Dictionary
- gồm cặp key:value
- {}
- có index
- 

```python
# VD
SinhVien = {"hoten":"Nguyen Van A", "diemtb":"9.0"}

print(SinhVien["hoten"])
```

```python
# truy cập bằng hàm get(key)
SinhVien.get("hoten")
```

```python
# thay đổi giá trị 
SinhVien['hoten'] = 'Vũ thị Linh'
print(SinhVien['hoten'])

# cách 2: sử dụng update truyền vô 1 dict
SinhVien.update({'hoten':'vũ thị lan'})
SinhVien.update({'hoten2':'vũ thị lan'})
print(SinhVien['hoten'])
```

```python
# thêm mục mới thì sd update nhưng với key mới
SinhVien.update({'noisinh':'Thái Nguyen'})
print(SinhVien)
```

```python
# xóa 1 mục bằng pop(key)
SinhVien.pop('hoten2')
print(SinhVien)

# popitem() xóa mục xuối cùng
SinhVien.popitem()
print(SinhVien)

# del dict[key] -> có thể xóa mục chỉ định hoặc toàn bộ dict
del SinhVien['hoten']
print(SinhVien)

```

```python
# Xóa sạch 1 dict => clear()
SinhVien.clear()
print(SinhVien)
```

```python
SinhVien = {'1':'Nguyễn Văn Á', '2':'ABC', '3':'XYX'}

# duyệt qua các key của dict
for key in SinhVien.keys():
    print(key)

# duyệt qua các values của dict:
for value in SinhVien.values():
    print(value)

# duyệt qua cặp key: value của dict
for key,value in SinhVien.items():
    print(key, value)
    
```

# BÀI TẬP VỀ DICTIONARY
- Xây dựng một từ điển có các chức năng như sau:
- (+) người dùng chọn thông qua menu
- (+) thêm một từ vựng mới kèm nghĩa của từ đó
- (+) tra cứu nghĩa của một từ
- (+) cập nhật ý nghĩa cho một từ vựng
- (+) Cho phép người dùng xóa đi một từ trong từ điển
- (+) Cho phép người dùng xóa toàn bộ tự vừng
- (+) cho phép ng dùng in ra toàn bộ từ vựng
- (+) Cho phép ng dùng in ra toàn bộ từ didển theo cấu trúc từ vựng: ý nghĩa
- (+) kết thúc chương trình
- 

```python
MyDict = dict()

while(True):
    print("\n" + "="*5 + "Welcome to my dictionary!"+ "="*5 + "\n")
    print("Vui lòng lựa chọn chức năng!")
    print("1 - thêm một từ vựng mới kèm nghĩa của từ đó")
    print("2 - tra cứu nghĩa của một từ")
    print("3 - cập nhật ý nghĩa cho một từ vựng")
    print("4 - xóa đi một từ")
    print("5 - in ra toàn bộ từ vựng")
    print("6 - in ra toàn bộ từ điển")
    print("7 - xóa toàn bộ từ điển")
    print("8 - Kết thúc chương trình")
    try:
        choose = int(input("Nhập lựa chọn: "))
    except ValueError:
        print(">> Vui lòng nhập số, không nhập chữ!")
        continue
    

    match choose:
        case 1:
            key = input("Nhập từ vựng: ")
            value = input("Nhập ý nghĩa: ")
            MyDict.update({key:value})
        case 2: 
            key = input("Từ cần tra: ")
            if key in MyDict.keys():
                print(MyDict[key])
            else:
                print("từ điển chưa có từ này!")
        case 3:
            key = input("Từ cần cập nhật ý nghĩa: ")
            if key in MyDict:
                new_value = input("Ý nghĩa mới: ")
                MyDict.update({key:new_value})
            else:
                print("Từ điển k có từ này!")
                
        case 4:
            key = input("Từ cần xóa: ")
            MyDict.pop(key)
        case 5:
            for key in MyDict.keys():
                print(key)
        case 6:
            for key, value in MyDict.items():
                print(key, ":", value)
        case 7:
            MyDict.clear()
            print("Từ điển đã được xóa nội dung")
        case 8:
            print("Chương trình kết thúc")
            break
        case _:
            print("Lựa chọn không hợp lệ, mời nhập lại")
            continue
            
```
