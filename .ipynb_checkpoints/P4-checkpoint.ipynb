# Sự khác biệt giữa list và tuple
- List có thể cập nhật, xóa, thay đổi,... Mutable -> []
- Tuple thì k thể thay đổi đc (Immutable) -> () nằm trong ngoặc tròn
- 

```python
GioiTinh = ('Nam','Nữ')

# Các thao tác với Tuple
## truy cập
print(GioiTinh[0])

for x in GioiTinh:
    print(x)

# Cộng (gộp tuple)
#  nhân tuple -> nó sẽ lặp lại các phần tử trong tuple
print('Nam' in GioiTinh)

# độ dài tuple
x = len(GioiTinh)
# đếm số lần xuất hiện
print(GioiTinh.count('Nữ'))

# min, max dùng đc cho cả dữ liệu string hoặc numberic
# sum chỉ dùng đc với dữ liệu sum nhé
print(max(GioiTinh))
print(min(GioiTinh))

num = (1,2,3)
print(sum(num))

num2 = sorted(num, reverse = True)
print(num2)
```

# set
- Là một tập hợp không có thứ tự, k thể thay đổi immutable và không đc lập chỉ mục. Lưu ý: Các mục set là k thể thay đôi nhưng bạn có thể xóa các mục hoặc thêm các mục mới
-  giá trị k đc trùng nhau -> {}

```python
monHoc = {'Toán', 'Lý'}
print(monHoc)
```

```python
# Duyệt các phần tử của set
for i in monHoc:
    print(i)
```

```python
# thêm một phần tử vào set:
# (1) Dùng add
monHoc.add('Lịch Sử')
print(monHoc)

# (2) dùng update
monHoc.update({"Anh Văn", "Đàn Piana"})
print(monHoc)

# khi thêm list vào set thì chỉ để lại 1 giá trị duy nhất nhé -> k lặp lại
```

```python
# xóa phần tử:
# monHoc.remove("Toán")
# print(monHoc)

# sử dụng discard (có thì xóa, k có thì k lỗi, còn remove nếu k tìm thấy thì sẽ báo lỗi)
monHoc.discard("Lý")
print(monHoc)
```

```python
# xóa phần tử đầu tiên -> pop
monHoc.pop()
print(monHoc)
```

```python
# xóa sạch clear()
monHoc.clear()
print(monHoc)
```

```python
# xóa hẳn set đi -> del
del(monHoc)
print(monHoc)
```

# Bài tập rút thăm trúng thưởng
- Xây dựng chương trình rút thăm trúng thưởng với các chức năng sau:
- (+) thêm một mã số dự thưởng vào thùng
- (+) Xóa một mã số dự thưởng ra khỏi thùng
- (+) Quay số ngẫy nhiên lấy ra một mã số trúng thưởng => dùng random nhé
- 

```python
import random 
# khai báo thùng phiếu
thung = set()

while(True):
    print('-----------')
    print("Vui lòng lựa chọn chức năng!")
    print("1 - thêm một mã dự thưởng vào thùng")
    print("2 - xóa một mã dự thưởng ra khỏi thùng")
    print("3 - Quay số ngẫu nhiên lấy ra một mã số trúng thưởng")
    print("4 - Kết thúc chương trình")
    luaChon = int(input())
    if (luaChon == 1):
        sdt = input("Nhap vao sđt: ")
        thung.add(sdt)
    elif(luaChon == 2):
        delsdt = input("Nhap sđt cần xóa: ")4
        thung.discard(delsdt)
    elif(luaChon == 3):
        if len(thung) == 0:
            print("Thùng rỗng!")
        else:
            winner = random.choice(list(thung))
            print(f"Chúc mừng {winner} đã trúng thưởng")
            thung.discard(winner)
    elif(luaChon == 4):
        print("Thoat")
        break
    else:
        break
    
        
    
```
