# Kiểu dữ liệu
- Trình thông dịch tự đoán kiểu dữ liệu
- int: số nguyên
- float: số thực
- complex: số phức (1+2j)
- bool: True/False
- str: Chuỗi ký tự
- NoneType: Chỉ ra giá trị null

```python
x = None
print(type(x))
```

```python
x= 5
print(type(x))
```

```python
x=5.0
print(type(x))
```

```python
x = 2+5j
print(type(x))
```

```python
x= 'qưểtyui'
print(type(x))
```

```python
x= True
print(type(x))
```

# Các phép toán tử trong python
- Cộng: +
- Trừ : -
- Nhân: *
- Chia: /
- Chia lấy dư: %
- Mũ: **
- Chia lấy nguyên: //

# Toán tử so sánh
- lớn hơn: >
- nhỏ hơn： <
- bằng: ==
- khác: !=
- lớn hơn hoặc bằng: >=
- nhỏ hơn hoặc bằng <=

# toán tử logic
- and: tất cả phải thỏa mãn
- or: ít nhất 1 cái thỏa mãn
- not: đảo chiều logic

```python
x = 5
y = 10.5
print(x<y)
```

```python
print(x>y)
```

```python
print(x>=y)
```

```python
print(x<=y)
```

```python
print(x!=y)
```

# Toán tử bit (thao tác trên bit kiểu 1101 đó, python sẽ nhìn dữ liệu kiểu dạng bit rồi mới tính)
- & giống như and  --> a&b
- | giống or --> a|b
- ^ giống XOR --> a^b
- ~ giống not --> ~b
- << giống left shift --> a<<n
- '>> giống right shift --> a>>n
